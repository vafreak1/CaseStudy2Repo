```python
import numpy as np
import matplotlib as mp
import glob
import os 

from matplotlib import pyplot as plt 

pwd = os.getcwd()


numpy_vars = {}
for np_name in glob.glob(str(pwd) + '/Data/masks/*.np[yz]'):
    numpy_vars[np_name] = np.load(np_name)

def save_coordinates(array):
    return (np.array(np.where(array != 0))[1:3])

coordinates_dict = {}
for filename, array in numpy_vars.items():
    coordinates_x = save_coordinates(array)[0]
    coordinates_y = save_coordinates(array)[1]
    coordinates = list(zip(coordinates_x, coordinates_y))
    coordinates_dict[filename] = coordinates

# Print the dictionary to verify the results
for filename, coordinates in coordinates_dict.items():
    print(f"Filename: {filename}")
    for x, y in coordinates:
        print(f"Coordinates: ({x}, {y})")
    print()
```
